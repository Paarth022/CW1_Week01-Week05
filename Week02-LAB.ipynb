5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks â‰¥ 50) and failed.

````python

#NAME:Paarth Baphna
#MISIS:M01095290

#Marks 
l=[86,59,75,96,74,78,91,68,41,63]
s=0
c=0
for i in range(0,len(l)-1):
    s=s+l[i]
    c=c+1
avg=s/c
hi=max(l)
lo=min(l)
c1=0
for marks in l:
    if marks>=50:
      c1=c1+1
c2=len(l)-c1
print(f"Average marks={avg}")
print(f"Highest marks ={hi}")
print(f"lowest marks={lo}")
print(f"Number of students who passed ={c1}")
print(f"Number of students who failed ={c2}")
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
#NAME:Paarth Baphna
#MISIS:M01095290

#import random library
import random

# Generates random number between 1 and 100
secret_number = random.randint(1, 100)

tries = 0
guess = None

print("I'm thinking of a number between 1 and 100.")
print("Try to guess it!")
while guess != secret_number:
    guess = int(input("Enter your guess: "))
    tries += 1

    if guess < secret_number:
        print("Too small! Try again.")
    elif guess > secret_number:
        print("Too large! Try again.")
    else:
        print(f"Correct! The secret number was {secret_number}.")
        print(f"You guessed it in {tries} tries!")
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each studentâ€™s score and displays the highest and second highest scores.


````python
#NAME:Paarth Baphna
#MISIS:M01095290

# Ask for number of students
num_students = int(input("Enter the number of students: "))

#highest and second highest scores
highest = -1
second_highest = -1

for i in range(num_students):
    score = float(input(f"Enter score for student {i + 1}: "))
    if score > highest:
        second_highest = highest
        highest = score
    elif score > second_highest:
        second_highest = score

#result
print(f"\nHighest score: {highest}")
print(f"Second highest score: {second_highest}")
````
