3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
#NAME:Paarth Baphna
#MISIS:M01095290

def password_to_ascii(password):
# convert each char to ASCII and join as string
    ascii_codes = [str(ord(char)) for char in password]
    return ''.join(ascii_codes)

user_password = input("Enter your password: ")  
ascii_password = password_to_ascii(user_password)
print("Password as ASCII string:", ascii_password)  
````

12. Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

````python
#NAME:Paarth Baphna
#MISIS:M01095290

def is_valid_password(pwd):
    if ' ' in pwd:
        print("Error: Password should not contain spaces.")  
        return False
    
# check all conditions
    has_length = len(pwd) > 8
    has_upper = any(c.isupper() for c in pwd)
    has_lower = any(c.islower() for c in pwd)
    has_digit = any(c.isdigit() for c in pwd)
    has_special = any(c in '@#$%^&*!' for c in pwd)
    
    if has_length and has_upper and has_lower and has_digit and has_special:
        print("Password is valid.")
        return True
    else:
        print("Invalid password.")
        
        return False

password = input("Enter your password: ") 

# check until valid
while not is_valid_password(password):
    password = input("Enter your password again: ")
````

13. Secret Word Generator with Randomization: Ask the user for their first name and surname. Create a function that generates a special secret word that:
Uses the first 3 letters of the first name and last 2 letters of the surname.
Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

````python
#NAME:Paarth Baphna
#MISIS:M01095290

import random

def generate_secret_word(first_name, surname):
   
    part1 = first_name[:3]
  
    part2 = surname[-2:]
    
    special_chars = "@#$%^&*!"
    random_char = random.choice(special_chars)

    secret_word = part1 + part2 + random_char
    return secret_word
# get name nd surname
fname = input("Enter your first name: ")   
sname = input("Enter  your surname: ")       
word = generate_secret_word(fname, sname)
print("Your secret word is:", word)    
````

15. (Binary to decimal) Write a function that parses a binary number as a string and convert it into a decimal integer. Use the function header: def binaryToDecimal(binaryString): For example, binary string 10001 is 17 (1 24 + 0
23 + 0 22 + 0 21 + 1 * 20 = 17)

So, binaryToDecimal("10001") returns 17.

Write a test program that prompts the user to enter a binary string and displays the corresponding decimal integer value.

````python
#NAME:Paarth Baphna
#MISIS:M01095290

def binaryToDecimal(binaryString):
    decimal = 0
    length = len(binaryString)
    for i in range(length):
# convert each bit to int and calculate value
        bit = int(binaryString[i])
        power = length - i - 1
        decimal += bit * (2 ** power)
    return decimal

#input
binary_input = input("Enter a binary number: ")
print("Decimal value:", binaryToDecimal(binary_input))
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
#NAME:Paarth Baphna
#MISIS:M01095290

def binaryToHex(binaryValue):
# convert binary string to decimal first
    decimal = 0
    length = len(binaryValue)
    for i in range(length):
        bit = int(binaryValue[i])
        power = length - i - 1
        decimal += bit * (2 ** power)
# convert decimal to hex
    hex_value = hex(decimal)[2:]  
    return hex_value

# input test
binary_input = input("Enter a binary number: ")
print("Hexadecimal value:", binaryToHex(binary_input))
````

17. (Decimal to binary) Write a function that converts a decimal number into a binary number as a string. Use the function header: def decimalToBinary(value):
Write a test program that prompts the user to enter a decimal integer value and displays the corresponding binary value.

````python
#NAME:Paarth Baphna
#MISIS:M01095290

def decimalToBinary(value):
    if value == 0:
        return "0"
    binary_str = ""
    while value > 0:
        remainder = value % 2
        binary_str = str(remainder) + binary_str
        value = value // 2
    return binary_str
decimal_input = int(input("Enter a decimal integer: "))
print("Binary value:", decimalToBinary(decimal_input))
````
