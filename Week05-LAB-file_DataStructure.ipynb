### Activity 13: Identifying Duplicate Entries in Clothing Stock

Create a program that reads the `clothing_stock.txt` file and identifies if there are any duplicate entries (both item and size). Use a set to find and print these duplicates.

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

`````python
#NAME:Paarth Baphna
#MISIS:M01095290

with open("clothing_stock.txt", "r") as file:
    items = [line.strip() for line in file.readlines()]
seen = set()
duplicates = set()

for item in items:
    if item in seen:
        duplicates.add(item)
    seen.add(item)

if duplicates:
    print("Duplicates found:")
    for dup in duplicates:
        print(f"  {dup}")
else:
    print("No duplicates found")
`````

### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python
#NAME:Paarth Baphna
#MISIS:M01095290

from collections import defaultdict

size_count = defaultdict(set)

with open("clothing_stock.txt", "r") as file:
    for line in file:
        item, size = line.strip().split(", ")
        size_count[item].add(size)
result = {item: len(sizes) for item, sizes in size_count.items()}
print(result)


for item, sizes in size_count.items():
    print(f"{item}: {len(sizes)} different sizes - {sizes}")
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
#NAME:Paarth Baphna
#MISIS:M01095290

study_groups = {
    "John": ["Math", "Physics"],
    "Emily": ["Physics", "Biology"],
    "Alex": ["Math"]
}
def add_group(student, group):
    if student not in study_groups:
        study_groups[student] = []
    if group not in study_groups[student]:
        study_groups[student].append(group)

def remove_group(student, group):
    if student in study_groups and group in study_groups[student]:
        study_groups[student].remove(group)

def show_groups(student):
    if student in study_groups:
        return study_groups[student]
    return []

add_group("John", "Chemistry")
print(f"John's groups: {show_groups('John')}")
remove_group("Emily", "Physics")
print(f"Emily's updated groups: {show_groups('Emily')}")
`````

### Activity 16: Managing and Shuffling Playlists

You are developing a music player that lets users create a playlist and shuffle it. Users should be able to add songs, remove songs, and shuffle the playlist randomly. What data structure would you use to represent the playlist, and why?

**Example data:**

```python
["Song A", "Song B", "Song C", "Song D"]
````

Steps:

1. Choose a data structure that allows easy addition, removal, and shuffling of songs.
2. Write a program that simulates playlist operations (add, remove, shuffle).
3. Let users shuffle the playlist and display the new order.
<hr>

`````python
#NAME:Paarth Baphna
#MISIS:M01095290

import random

class Playlist:
    def __init__(self):
        self.songs = []
    
    def add_song(self, song):
        self.songs.append(song)
        print(f"Added: {song}")
    
    def remove_song(self, song):
        if song in self.songs:
            self.songs.remove(song)
            print(f"Removed: {song}")
    
    def shuffle(self):
        random.shuffle(self.songs)
        print("Playlist shuffled!")
    
    def display(self):
        for i, song in enumerate(self.songs, 1):
            print(f"{i}. {song}")
playlist = Playlist()
playlist.add_song("Song A")
playlist.add_song("Song B")
playlist.add_song("Song C")
playlist.shuffle()
playlist.display()
`````

3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

`````python
#NAME:Paarth Baphna
#MISIS:M01095290

from collections import Counter
import string

with open("Les_Brown.txt") as f:
    words = f.read().lower().translate(str.maketrans('', '', string.punctuation)).split()

counts = Counter(words)

with open("word_count.txt", "w") as f:
    f.writelines([print(f"{w}: {c}") or f"{w}: {c}\n" for w, c in counts.items()])
`````

### Activity 20: Managing Event Tickets

You are developing a system to manage event tickets for concerts or sports events. Each event has a limited number of tickets available, and users can purchase tickets if they are still available. You need to track which users have purchased tickets for specific events and how many are still available. What data structure would you use for this task, and why?

**Example data:**
- Available tickets: `{"Concert": 100, "Football Game": 50, "Theater Show": 75}`
- Purchasers: `{"Concert": ["John", "Emily"], "Football Game": ["Alex"], "Theater Show": ["Sam", "John"]}`

**Steps:**
1. Choose a data structure to track ticket availability and purchasers efficiently.
2. Write a program that handles ticket purchases and updates the list of purchasers.
3. Implement functions to check availability and view who has purchased tickets.
<hr>

`````python
#NAME:Paarth Baphna
#MISIS:M01095290

class TicketSystem:
    def __init__(self):
        self.available = {"Concert": 100, "Football Game": 50, "Theater Show": 75}
        self.purchasers = {"Concert": [], "Football Game": [], "Theater Show": []}
    
    def purchase_ticket(self, event, user):
        if event in self.available and self.available[event] > 0:
            self.available[event] -= 1
            self.purchasers[event].append(user)
            print(f"{user} purchased ticket for {event}")
            return True
        print(f"No tickets available for {event}")
        return False
    
    def check_availability(self, event):
        return self.available.get(event, 0)
    
    def show_purchasers(self, event):
        return self.purchasers.get(event, [])
system = TicketSystem()
system.purchase_ticket("Concert", "John")
print(f"Concert tickets left: {system.check_availability('Concert')}")
print(f"Purchasers: {system.show_purchasers('Concert')}")
`````

### Activity 21: Personal Finance App

You are creating a personal finance app that allows users to categorize their expenses and track spending by category (e.g., food, entertainment, travel). Users can add expenses, remove expenses, and see a summary of their spending. What data structure would you choose to store and organize the expenses, and why?

**Example data:**

```python
{"Food": [25, 40, 15], "Entertainment": [50, 20], "Travel": [300, 150]}
```
Steps:

1. Choose a data structure that lets users easily add and track expenses by category.
2. Write a program that allows users to input expenses, remove them, and view their total spending by category.
3. Implement a function to display a summary of all spending.
<hr>

`````python
#NAME:Paarth Baphna
#MISIS:M01095290

class FinanceApp:
    def __init__(self):
        self.expenses = {
            "Food": [],
            "Entertainment": [],
            "Travel": []
        }
    
    def add_expense(self, category, amount):
        if category in self.expenses:
            self.expenses[category].append(amount)
            print(f"Added ${amount} to {category}")
    
    def remove_expense(self, category, amount):
        if category in self.expenses and amount in self.expenses[category]:
            self.expenses[category].remove(amount)
            print(f"Removed ${amount} from {category}")
    
    def category_total(self, category):
        return sum(self.expenses.get(category, []))
    
    def summary(self):
        for category, amounts in self.expenses.items():
            total = sum(amounts)
            print(f"{category}: ${total:.2f}")
app = FinanceApp()
app.add_expense("Food", 25)
app.add_expense("Food", 40)
app.add_expense("Entertainment", 50)
app.summary()
`````
